```python
# Step 1: Define the function for initializing the inventory with error handling
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory
```

````python
# Définir les produits disponibles
products = ["t-shirt", "mug", "hat", "book", "keychain"]

# Initialiser l'inventaire avec des quantités pour chaque produit
inventory = {}
for product in products:
    while True:
        quantity = input(f"Entrez la quantité disponible pour '{product}': ")
        try:
            quantity = int(quantity)
            if quantity < 0:
                print("La quantité doit être un nombre non négatif.")
            else:
                inventory[product] = quantity
                break
        except ValueError:
            print("Veuillez entrer un nombre entier.")

# Fonction pour saisir les commandes des clients
def saisir_commandes():
    customer_orders = set()  # Initialiser l'ensemble des commandes clients
    
    while True:
        product = input("Nom du produit que vous souhaitez commander : ")
        if product in products:
            while True:
                quantity = input(f"Quantité de '{product}' que vous souhaitez commander : ")
                try:
                    quantity = int(quantity)
                    if quantity < 0:
                        print("La quantité doit être un nombre positif.")
                    else:
                        customer_orders.add((product, quantity))  # Ajouter la commande à l'ensemble des commandes
                        break
                except ValueError:
                    print("Veuillez entrer un nombre entier pour la quantité.")
        else:
            print("Le produit saisi n'est pas disponible.")
        
        another_order = input("Voulez-vous commander un autre produit ? (oui/non) ")
        if another_order.lower() != 'oui':
            break  # Sortir de la boucle si l'utilisateur ne souhaite plus commander
    
    return customer_orders

# Fonction pour calculer le prix total des commandes clients
def calculate_total_price(customer_orders):
    total_price = 0
    for product, quantity in customer_orders:
        while True:
            try:
                price = float(input(f"Entrez le prix unitaire pour '{product}': "))
                if price < 0:
                    print("Le prix doit être un nombre non négatif.")
                else:
                    total_price += price * quantity
                    break
            except ValueError:
                print("Veuillez entrer un nombre valide.")
    return total_price

# Fonction pour mettre à jour l'inventaire et filtrer les produits épuisés
def mettre_a_jour_inventaire(inventory, customer_orders):
    updated_inventory = {product: inventory[product] - sum(quantity for p, quantity in customer_orders if p == product)
                         for product in inventory}
    updated_inventory = {product: quantity for product, quantity in updated_inventory.items() if quantity > 0}
    return updated_inventory

# Appel de la fonction pour saisir les commandes
customer_orders = saisir_commandes()

# Mise à jour de l'inventaire
inventory = mettre_a_jour_inventaire(inventory, customer_orders)

# Calcul du prix total des commandes clients
total_price = calculate_total_price(customer_orders)

# Affichage de l'inventaire mis à jour
print("Inventaire mis à jour :")
for product, quantity in inventory.items():
    print(f"{product}: {quantity}")

# Affichage du prix total des commandes
print(f"Prix total des commandes : {total_price:.2f}")
````
